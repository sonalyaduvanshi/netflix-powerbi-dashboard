```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {'title': ['Movie1', 'Movie2', 'Movie3'],
        'listed_in': ['Action, Drama', 'Comedy, Drama', 'Action'],
        'rating': [4.5, 3.8, 4.2]}

df = pd.DataFrame(data)

df_exploded = df['listed_in'].str.split(',').explode()

genre_counts = df_exploded.value_counts()

plt.figure(figsize=(5,4))  
sns.barplot(x=genre_counts.index, y=genre_counts.values)  
plt.title('Genre Distribution')  
plt.xlabel('Genres') 
plt.ylabel('Count')  
plt.xticks(rotation=40)  
plt.show() 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df_numeric = df.select_dtypes(include=['number'])

df_numeric = df_numeric.fillna(df_numeric.mean())  

corr = df_numeric.corr()
plt.figure(figsize=(8, 4))  
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.histplot(df['rating'], kde=True, bins=20, color='blue')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df, palette='viridis')
plt.title('Movies vs TV Shows')
plt.xlabel('Type (Movie/TV Show)')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_exploded = df['listed_in'].str.split(',').explode()
plt.figure(figsize=(6, 5))
sns.countplot(y=df_exploded, hue=df_exploded, palette='coolwarm', legend=False)
plt.title('Genre Distribution')
plt.xlabel('Count')
plt.ylabel('Genres')
plt.show()
```
